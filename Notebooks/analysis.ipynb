```python
import pandas as pd
import matplotlib.pyplot as plt
print("Project Started")
```

```python
orders = pd.read_csv(r'C:\Users\ASUS\Desktop\Ecommerce-Analysis\Data\orders.csv')
customers = pd.read_csv(r'C:\Users\ASUS\Desktop\Ecommerce-Analysis\Data\customers.csv')

print(orders.head())
print(customers.head())
```

```python
print(orders.shape)
print(customers.shape)
```

```python
print(f"Total Orders: {orders.shape[0]}")
print(f"Total Customers: {customers.shape[0]}")
print(f"Total Revenue: {orders['amount'].sum()}")
```

```python
max_purchaser=orders.groupby('customer_id')['amount'].sum().sort_values(ascending=False).head(1)
print(f"Customer with the highest purchase: {max_purchaser.index[0]} with amount {max_purchaser.values[0]}")
```

```python
merged_data = pd.merge(orders, customers, on='customer_id')
merged_data.groupby('city')['amount'].sum().sort_values(ascending=False)
city_revenue = merged_data.groupby('city')['amount'].sum().sort_values(ascending=False)

plt.figure()
city_revenue.plot(kind='bar')
plt.title("Revenue by City")
plt.xlabel("City")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['month'] = orders['order_date'].dt.to_period('M')
monthly_revenue = orders.groupby('month')['amount'].sum()
plt.figure()
monthly_revenue.plot(kind='line', marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
customer_spend = orders.groupby('customer_id')['amount'].sum()

plt.figure()
customer_spend.plot(kind='pie', autopct='%1.1f%%')
plt.title("Customer Contribution")
plt.ylabel("")
plt.show()
```
